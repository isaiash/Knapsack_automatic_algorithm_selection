```python
import os
import numpy as np
```

```python
PATH = 'outputs/'
for file in os.listdir(PATH):
    if file[:5] == 'cplex':
        f = open(PATH+str(file),'r')
        f2 = open(PATH+str(file)+'2','w')
        all_ = []
        anterior = [1, 99999,1]
        for line in f:
            a = line.split()
            a = [int(a[0]), float(a[1]), float(a[2])]
            
            if (anterior[0] == a[0] and anterior[1] > a[1]):
                all_.pop(-1)
            all_.append(a)
            anterior = a
        for l in all_:
            print(*l, file=f2)
```
